```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Struktur untuk menyimpan data film Ultraman
class Ultraman:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

# Bubble Sort Iteratif
def bubble_sort_iterative(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j].rating > arr[j + 1].rating:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Bubble Sort Rekursif
def bubble_sort_recursive(arr, n=None):
    if n is None:
        n = len(arr)

    # Base case
    if n == 1:
        return

    # Bubble up the largest element
    for i in range(n - 1):
        if arr[i].rating > arr[i + 1].rating:
            arr[i], arr[i + 1] = arr[i + 1], arr[i]

    # Recursive call
    bubble_sort_recursive(arr, n - 1)

# Grafik untuk menyimpan data
n_values = []
iterative_times = []
recursive_times = []

# Fungsi untuk memperbarui grafik
def update_graph():
    plt.figure(figsize=(8, 6))
    plt.plot(n_values, iterative_times, label='Iterative', marker='o', linestyle='-')
    plt.plot(n_values, recursive_times, label='Recursive', marker='o', linestyle='-')
    plt.title('Performance Comparison: Iterative vs Recursive')
    plt.xlabel('Number of Films (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table():
    table = PrettyTable()
    table.field_names = ["n", "Iterative Time (s)", "Recursive Time (s)"]
    min_len = min(len(n_values), len(iterative_times), len(recursive_times))
    for i in range(min_len):
        table.add_row([n_values[i], iterative_times[i], recursive_times[i]])
    print(table)

# Program utama
if __name__ == "__main__":
    film_data = [
        Ultraman("Ultraman Trigger", 6.9),
        Ultraman("Ultraman:The Ultimate Hero", 6.6),
        Ultraman("Ultraman Neos", 7.4),
        Ultraman("Ultraman Blazar", 8.1),
        Ultraman("Ultraman Ginga S", 6.9),
        Ultraman("Ultraman Orb", 8.2),
        Ultraman("Ultraman X", 7.7),
        Ultraman("Ultraman Trigger", 6.9),
        Ultraman("Ultraman:The Ultimate Hero", 6.6),
        Ultraman("Ultraman Neos", 7.4),
        Ultraman("Ultraman Blazar", 8.1),
        Ultraman("Ultraman Ginga S", 6.9),
        Ultraman("Ultraman Orb", 8.2),
        Ultraman("Ultraman X", 7.7),
        Ultraman("Ultraman Tiga", 8.0),
        Ultraman("Ultraman Dyna", 7.9),
        Ultraman("Ultraman Gaia", 8.4),
        Ultraman("Ultraman Geed", 8.0),
        Ultraman("Ultraman Mebius", 8.1),
        Ultraman("Ultraman Z", 8.0),
        Ultraman("Ultraman Max", 7.2),
        Ultraman("Ultraman Cosmos", 7.8),
        Ultraman("Ultraman Decker", 7.1),
        Ultraman("Ultraman R/B", 7.8),
        Ultraman("Ultraman Taiga", 7.1),
        Ultraman("Ultraman Nexus", 8.4),
        Ultraman("Ultraseven", 8.2),
    ]

    while True:
        try:
            n = int(input("Masukkan jumlah film (atau ketik -1 untuk keluar): "))
            if n == -1:
                print("Program selesai. Terima kasih!")
                break

            if n < 1 or n > len(film_data):
                print(f"Masukkan nilai antara 1 dan {len(film_data)}!")
                continue

            current_data = film_data[:n]
            n_values.append(n)

            # Ukur waktu eksekusi Bubble Sort Iteratif
            start_time = time.time()
            bubble_sort_iterative(current_data[:])  # Copy untuk menjaga data asli
            iterative_times.append(time.time() - start_time)

            # Ukur waktu eksekusi Bubble Sort Rekursif
            start_time = time.time()
            bubble_sort_recursive(current_data[:])  # Copy untuk menjaga data asli
            recursive_times.append(time.time() - start_time)

            # Cetak tabel waktu eksekusi
            print_execution_table()

            # Perbarui grafik
            update_graph()

        except ValueError:
            print("Masukkan jumlah film yang valid!")
```
